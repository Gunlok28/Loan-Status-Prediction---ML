```python
for i in TD.columns:
    if TD[i].dtypes==object:
        TD[i] = TD[i].fillna(TD[i].mode()[0])
    else:
        TD[i] = TD[i].fillna(TD[i].mean())
```

```python
TD.isna().sum()
```

```python
X = TD.drop(labels=["Loan Status"],axis=1)
Y = TD[["Loan Status"]]
```
